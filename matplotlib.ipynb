Stack Plot

```python
days = np.arange(1,6)

sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]


plt.plot([],[], color='m', label='Sleeping', linewidth=5)
plt.plot([],[], color='c', label='Eating', linewidth=5)
plt.plot([],[], color='r', label='Working', linewidth=5)
plt.plot([],[], color='k', label='Playing', linewidth=5)

plt.stackplot(days, sleeping, working,playing, colors=['m','c','r','k'])

plt.xlabel('X')
plt.ylabel('y')
plt.title('Stack Plot')
plt.legend()
plt.show()
```

Pie Chart

```python
slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','b']

plt.pie(slices,
        labels=activities,
        colors=cols,
        startangle=90, #where to start drawing
        shadow=True,
        explode=(0,0,0.1,0), #puling one of the slices out
        autopct ='%1.1f%%')
plt.title('Pie Plot')
plt.show()
```
